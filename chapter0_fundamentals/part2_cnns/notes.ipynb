[Six (and a half) intuitions for KL divergence](https://www.lesswrong.com/posts/no5jDTut5Byjqb4j5/six-and-a-half-intuitions-for-kl-divergence) Notes:

1. Expected Surprise:
* **Surprise = −log(probability)** measures how wrong or shocked you are when an event happens; assigning low probability to something that occurs leads to a large penalty.
* Entropy = average surprise under the true distribution; it measures how unpredictable the world is by averaging how surprised you are across all possible outcomes.
* Expected surprise under Q but averaged over P measures how surprised your model is by reality; it is high when your predicted probabilities don’t match what actually happens.
    * Expected value = weighted average over outcomes, where weights are how often each outcome actually occurs (the true distribution).
* KL divergence = extra average surprise you experience when using a wrong model Q instead of the true distribution P.
    * KL stands for Kullback-Leibler, named after Solomon Kullback and Richard Leibler.
    * It's called a divergence because it measures how much one probability distribution $Q$ differs from another $P$, but it's not a true "distance" since it's not symmetric: $D_{KL}(P \| Q) \neq D_{KL}(Q \| P)$
* KL is not symmetric because it only penalizes errors on events that actually occur (from P), so being wrong about common events is much worse than being wrong about rare ones.

2. Hypothesis Testing:
* We compare two hypotheses: $H_0: X \sim P$ (true) vs $H_1: X \sim Q$.
* Use the likelihood ratio $\frac{p_x}{q_x}$ (or $\ln p_x - \ln q_x$) to decide which hypothesis an observation supports.
* $\ln p_x - \ln q_x$ represents the evidence a single observation provides for $P$ over $Q$.
* $D_{KL}(P \| Q) = \mathbb{E}_P[\ln p_x - \ln q_x]$: KL divergence is the average evidence per sample in favor of $P$ over $Q$.
* If $P$ and $Q$ are similar → little evidence → KL is small; if very different → strong evidence → KL is large.

3. MLE
* Training a model = making its predictions match reality by reducing how surprised it is by real data
* Maximum likelihood training is equivalent to minimizing KL divergence (minimizing cross-entropy), meaning we train models by making their predicted distribution Q as close as possible to the true data distribution P.

4. Suboptimal Coding:
* Shorter codes for common events minimize average message length because we optimize for what happens most often
* Being wrong about probabilities = wasting communication resources
* KL divergence measures inefficiency: how many extra bits you waste by using the wrong distribution Q instead of the true distribution P
