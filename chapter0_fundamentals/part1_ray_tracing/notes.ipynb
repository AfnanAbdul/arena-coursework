* **Use `torch.where` for elementwise conditional selection instead of boolean indexing when the replacement depends on another tensor.** Boolean indexing (e.g., `out[mask] = s`) flattens the selected elements and can cause shape mismatch errors. `t.where(mask, s, inf)` works elementwise and preserves the full tensor shape, which is ideal when building a `(nrays, ntriangles)` tensor of candidate distances (valid hit → `s`, invalid → `inf`) before taking the minimum across triangles.
    - `torch.where(condition, value_if_true, value_if_false)`

* **Get in the habit of checking tensor shapes with `assert` during development.** After major transformations (repeat, broadcast, stack, solve), add assertions like `assert mat.shape == (NR, NT, 3, 3)` or `assert s.shape == (NR, NT)` to quickly catch mistakes in batching logic.

* **Use type annotations with shape hints (e.g., `Float[Tensor, "NR 3 3"]`) to document tensor structure.** This makes code easier to reason about and helps ensure functions receive tensors with the expected dimensions (especially useful in batched linear algebra or ray–triangle intersection code).

* **Prefer repeating the full tensor and then unpacking instead of repeating each component separately.** For example, repeat the whole `triangles` tensor once and then unpack (`A, B, C = triangles`), and do the same for rays (`O, D = rays`). This leads to cleaner code, fewer operations, and makes shape reasoning easier.
